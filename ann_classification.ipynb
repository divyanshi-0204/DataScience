```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv('Churn_Modelling.csv')
data.head()
```

```python
X = data.iloc[:,3:-1].values
print(X)
```

```python
y = data.iloc[:,-1].values
print(y)
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
X[:,2] = np.array(le.fit_transform(X[:,2]))
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer([('encoder',OneHotEncoder(),[1])],remainder='passthrough')
X = np.array(ct.fit_transform(X))
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
import tensorflow as tf
```

```python
ann = tf.keras.models.Sequential()
```

```python
#creating hidden layers
ann.add(tf.keras.layers.Dense(units=6,activation="relu"))
#units=no. of neurons
#activation=specify which activation function is used
#relu=rectified linear unit
```

```python
ann.add(tf.keras.layers.Dense(units=6,activation="relu"))
```

```python
#output layer
ann.add(tf.keras.layers.Dense(units=1,activation="sigmoid"))
#for binary, output is 0 or 1, use sigmoid function and one neuron
```

```python
#compiling
ann.compile(optimizer="adam",loss="binary_crossentropy",metrics=["accuracy"])
```

```python
#fitting
ann.fit(X_train,y_train,batch_size=32,epochs = 100)#epchos=how many times neural network shpould be trained
```

```python
print(ann.predict(sc.transform([[1,0,0,600,1,40,3,60000,2,1,1,50000]]))>0.5)
```
